```python
#we can not take every user review and books to make our model so we make some criteria best on that we build our model
#1.select users who had given minimum 200 review and ratings
#2 consider the books which get minimum 50 ratings
```

```python
ratings['user_id'].value_counts()
```

```python
x=ratings['user_id'].value_counts()>200
x
```

```python
x[x].shape
```

```python
#this gives user_id of user which give more than 200 ratings
y=x[x].index
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head(4)
```

```python
ratings.shape
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books
```

```python
number_ratings=ratings_with_books.groupby('title')['ratings'].count().reset_index()
```

```python
number_ratings.rename(columns={'ratings':'number of ratings'},inplace=True)
```

```python
final_ratings=ratings_with_books.merge(number_ratings,on='title')
```

```python
final_ratings.head(5)
```

```python
final_ratings.shape
```

```python
final_ratings=final_ratings[final_ratings['number of ratings']>=50]
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_ratings.shape
```

```python
book_pivot=final_ratings.pivot_table(columns='user_id',index='title',values='ratings')
```

```python
book_pivot.shape
```

```python
book_pivot.head(5)
```

```python
#noe convert nan value in 0
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
#we use KNN clasifier and use nearest neighboring method to cluster nearest points
```

```python
#but we have 0's in df so we convert into sparse metrix so we use only non zero value for computation.
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
#we make a fucntion where we give books name and it will give books suggestions
```

```python
def recommend_book(book_name):
 book_id = np.where(book_pivot.index == book_name)[0][0]

 distance,suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

 for i in range(len(suggestions)):
  print('Based on your Book, suggestions are : ')
  print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
